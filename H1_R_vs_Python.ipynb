### 4. In Python, create a list prices = [10, 20, 30] . Try to divide it by 2 (i.e., prices /2). Does it work?

### 8. Based on your observations from the previous two exercises, what is the difference between the indexing and “stop” indices in R compared to Python?

With Python, indexing starts from 0 and stops at n - 1, where n represents the length of the list. With slicing in Python, we start from whatever idex we want (inclusive) and then end at the index + 1 we want to access (exclusive).

With R, indexing starts at 0 and ends at n, where n is the length of the vector. If we want to slice the vector, we start at where we want and end at the index we want (inclusive for both end points)
